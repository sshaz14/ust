```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings. filterwarnings('ignore')
```

```python
auto_price=pd.read_csv('Automobile_price_data__Raw_.csv', sep=',')
#display(auto_price)
```

```python
auto_price.drop(['symboling','normalized-losses'], axis=1, inplace=True)
```

```python
#Convert object columns to numerical columns
cols=['bore', 'stroke', 'horsepower','peak-rpm','price']
#args =('coerce')=> These columns have coersion of multiple dytype numeric+'?'
auto_price[cols]=auto_price[cols].apply(pd.to_numeric, args = ('coerce',))
```

```python
auto_price.isnull().sum()
```

```python
auto_price.dropna(inplace=True)
```

```python
auto_price['price'].plot(kind='line', figsize = (15,8))
```

```python
auto_price['price'].sort_values().unique()
```

```python
plt.style.use('seaborn-whitegrid')
plt.figure(figsize=(15,9))
plt.plot(auto_price['price'].sort_values().unique(), color = 'teal')
```

```python
plt.style.use('seaborn-whitegrid')
plt.figure(figsize=(15,9))
plt.plot(auto_price.sort_values(by='price')['price'],auto_price.sort_values(by='price')['city-mpg'] , color = 'lime')
plt.title('Price vi City Mileage Line Plot')
plt.xlabel('Price')
plt.ylabel('City Mileage')
```

```python
caty =auto_price['body-style'].unique()

plt.subplot(3, 2, 1)
sns.scatterplot(x = auto_price['highway-mpg'], y==auto_price['body-style'==caty[0]] ,marker = 's',alpha = 0.7, color = 'red')

plt.subplot(,2, 2)
sns.scatterplot(x = auto_price['highway-mpg'], y==auto_price['body-style'==caty[1]] ,marker = 's',alpha = 0.7, color = 'red')
```

```python
body_style=list(auto_price['body-style'].unique())
```

```python
plt.figure(figsize = (18,13))
for index, style in enumerate(body_style):
    plt.subplot(2, 3, index + 1)
    sns.scatterplot(x = auto_price[auto_price['body-style'] == style]['city-mpg'], 
            y = auto_price[auto_price['body-style'] == style]['price'], color = 'orangered', marker='s')
plt.show()
```

```python
colors_ = ['crimson','deeppink','lightgreen', 'deepskyblue','darkviolet']
marker_ = ['o','x','s', '*','.']

fig = plt.figure(figsize=(20,9))

for index, (style,label, mark) in enumerate (zip(body_style, colors_, marker_)):
    plt.subplot(2, 3, index + 1)
    sns.scatterplot(x = auto_price[auto_price['body-style'] == style]['city-mpg'], 
            y = auto_price[auto_price['body-style'] == style]['price'], color = label, marker= mark)
plt.show()
```
